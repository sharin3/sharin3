Image Enhancement techniques - Histogram

  Image enhancement is the process of adjusting digital images so that the results are more suitable for display or further image analysis[1]. Image histogram is the graphical representation of the intensity distribution of an image, it quatifies the number of pixel for each intensity value

  Histogram Equalization is used to improve the contrast in images by detecting the distribution of pixel densities in an image and plotting these pixel densities on a histogram. The distribution of this histogram is then examined and if there are ranges of pixel brightnesses that arent currently being used, the histogram is then stretched to cover those ranges, and then is back projected onto the image to increase the overall contrast of the image[2].
     Example., for obtaining the bright image from dark image we use histogram equalization. which is used to distribute the pixel equally and form the brighhter image.
     
     equalize_hist is used to distribute the image uniformly and helps to obtain clear image.

   **Scientific Paper**
    
     Early diagnosis of breast cancer using contrast limited adaptive histogram equalization (CLAHE) and Morphology methods 


     The aim of this paper is improve the hybrid solution for early detection of the breast cancer using mammogram images[4]. The standard histogram equalization is effective and simple method for contrast enhancement but for medical images most of the time it produce excessive contrast enhancement due to lack of control for the level of enhancement. In this paper "contrast limited adaptive histogram equalization"(CLAHE) is used. due to this the resulting image is free from noise. CLAHE has shown a good results in contrast enhancement of medical images.
     The CLAHE is a special type of adaptive histogram. This method divides the image into appropriate region and applies histogram equalization to them. here, it modifies the intensity values of the image in order to maximize the contrast for all pixel of the image. The new histogram obtained is different from normal Histogram[3].
     This method is used to enhance the finer detail of the mammogram images and gives the correct information which helps to diagonize the breast cancer.
