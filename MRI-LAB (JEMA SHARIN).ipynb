```python
%matplotlib inline
from skimage import novice, data, io, exposure
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import dicom

RefDs = dicom.read_file('IM_0003.dcm')
print(RefDs)
```

```python
RefDs = dicom.read_file('IM_0003.dcm')
print('The Pixel Spacing in file IM_0003 is '+str(RefDs.PixelSpacing))
print('The Echo Time in file IM_0003 is '+str(RefDs.EchoTime)+'ms')
print('The Image Position of the patient in file IM_0003 is '+str(RefDs.ImagePositionPatient))
print('The Scanning sequence the image in file IM_0003 is '+str(RefDs.ScanningSequence))
print('The name of the patient in file IM_0003 is '+str(RefDs.PatientsName))
print('The Acquisition Matrix in file IM_0003 is '+str(RefDs.AcquisitionMatrix))
print('The frequency in file IM_0003 is '+str(RefDs.ImagingFrequency)+'MHz')
print('The Flip Angle in file IM_0003 is '+str(RefDs.FlipAngle)+'degree')
print('The Repetition Time in file IM_0003 is '+str(RefDs.RepetitionTime)+'ms')
print('The image is '+str(RefDs.Rows)+'x'+str(RefDs.Columns)+' pixels')
```

```python
ConstPixelDims = (int(RefDs.Rows), int(RefDs.Columns))
ArrayDicom = np.zeros(ConstPixelDims, dtype=RefDs.pixel_array.dtype)
ds = dicom.read_file('IM_0003.dcm')
ArrayDicom[:,:]=ds.pixel_array

fig=plt.figure(figsize=(5, 5), dpi= 80, facecolor='w', edgecolor='k')
plt.axes().set_aspect('equal')
plt.set_cmap(plt.gray()) 
x=np.arange(RefDs.Rows)
y=np.arange(RefDs.Columns)

plt.pcolormesh(x, y, np.flipud(ArrayDicom[:, :]))
plt.axis([x.min(),x.max(),y.min(),y.max()])
```
